```
import numpy as np
import pandas as pd

from datetime import datetime

import matplotlib.pyplot as plt
%matplotlib inline
from google.colab import drive
drive.mount('/content/drive')
```

```
company_details_df = pd.read_csv('/content/company_sales_data.csv')
company_details_df.head(5)
```

# <u>**Problem 1**</u>

### Plot the total profit across the months

* X label name = Month Number
* Y label name = Total profit



```
# Your code here
plt.rcParams['figure.figsize'] = (10,5)
Profit = company_details_df['total_profit']
#create plot object
plt.plot(Profit)
plt.title('Total profit of company across months',fontsize=20)
plt.xlabel('Month Number',fontsize=20)
plt.ylabel('Total Profit',fontsize=20)
```

# <b><u>Problem 2 </u></b>

### Read face cream and facewash product sales data and show it using the bar chart. The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```
# Your code here
facecream_facewash = company_details_df[['facecream','facewash']]
facecream_facewash.plot(kind='bar')
plt.title('Sales Per Month For Facecream & Facewash',fontsize=18)
plt.xlabel('Units',fontsize=18)
plt.ylabel('Months',fontsize=18)
```

# <b><u>Problem 3</u></b>

### Display the number of units sold per month for each product using multiline plots. (i.e. separate plotline for each product ).

```
# Your code here
plt.plot(company_details_df['facecream'],label = 'Facecream Units Sold',linewidth = 4)
plt.plot(company_details_df['facewash'],label = 'Facewash Units Sold',linewidth = 4)
plt.plot(company_details_df['toothpaste'],label = 'Toothpaste Units Sold',linewidth = 4)
plt.plot(company_details_df['bathingsoap'],label = 'Bathingsoap Units Sold',linewidth = 4)
plt.plot(company_details_df['shampoo'],label = 'Shampoo Units Sold',linewidth = 4)
plt.plot(company_details_df['moisturizer'],label = 'Moisturizer Units Sold',linewidth = 4)

plt.title('The Number Of Units Sold Per Month For Each Product')
plt.xlabel('Months')
plt.ylabel('Products')
plt.legend(loc = 'upper left')
```
